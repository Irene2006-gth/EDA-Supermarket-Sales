## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart
df["Total"] = df["Unit price"] * df["Quantity"]
df.groupby("Product line")["Total"].sum().plot(kind="bar")
plt.title("Sales by Product Line")
plt.show()
````

##### 1. Why did you pick the specific chart?

This chart was chosen because it clearly shows the comparison between different product categories and helps in identifying patterns and differences in total sales across product lines.

##### 2. What is/are the insight(s) found from the chart?

Some product lines generate significantly higher total sales than others.
This indicates that certain products are more popular and in higher demand.
Other product lines show lower sales, suggesting weaker performance or lower customer interest.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes, the insights can create a positive business impact because:

The business can focus more on high-performing product lines to maximize revenue.
Marketing strategies can be designed to promote the most popular products.

However, there is also an indication of potential negative growth:

Product lines with low sales may reduce overall profitability.
This could be due to poor demand, pricing issues, or lack of promotion.
If not addressed, these underperforming categories can slow down business growth.
